**Exploratory Data Analysis (EDA)**

Exploratory Data Analysis (EDA) in data mining is a crucial step where initial investigations are conducted on datasets to discover patterns, spot anomalies, and test hypotheses, all before applying more complex machine learning models. EDA involves summarizing the main characteristics of the data, often using visual methods. It helps in understanding the structure of the data, detecting missing or erroneous values, and ensuring that the correct assumptions are met for further analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
```

```python
from google.colab import files
from os.path import exists
```

```python
# Task 1: Data Loading and Preliminary Inspection
# Load the Diabetes dataset

uploaded = files.upload()
```

```python

diabetes_data = load_diabetes()
df = pd.DataFrame(data=diabetes_data.data, columns=diabetes_data.feature_names)
df['Disease Progression'] = diabetes_data.target  # Adding the target variable

```

```python
# Task 2: Initial Data Overview
print("Shape of DataFrame:", df.shape)
print("First 5 rows of the DataFrame:")
print(df.head())
```

```python
print("Last 5 rows of the DataFrame:")
print(df.tail())
```

```python
# Task 3: Data Types and Missing Values
print("Data types and missing values:")
print(df.info())
```

```python
# Task 4: Descriptive Statistics
print("Statistical summary of the features and target:")
print(df.describe())
```

```python
# Task 5: Data Distribution Visualization
# Selecting a subset of features for histograms
features_for_histogram = ['bmi', 'bp']
df[features_for_histogram].hist(bins=15, figsize=(15, 5), layout=(1, 2))
plt.show()

```

```python
# Task 6: Simple Visualization of Relationships
# Scatter plots for 'BMI' vs 'Disease Progression' and 'BP' vs 'Disease Progression'
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x='bmi', y='Disease Progression', data=df)
plt.title('BMI vs Disease Progression')

```

```python
print(df.info())
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x='bp', y='Disease Progression', data=df)
plt.title('BP vs Disease Progression')
plt.show()



```

```python
# Select 3 columns for the pairplot
selected_columns = df[['bmi', 'bp', 's1']]

# Create a pairplot using seaborn
sns.pairplot(selected_columns)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Box Plot for Multiple Features
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['bmi', 'bp', 's1', 's2', 's3']])
plt.title('Box Plot of Selected Features')
plt.ylabel('Values')
plt.show()

```

```python
# Violin Plot for Selected Features
plt.figure(figsize=(10, 6))
sns.violinplot(data=df[['bmi', 'bp', 's1', 's2', 's3']])
plt.title('Violin Plot of Selected Features')
plt.ylabel('Values')
plt.show()
```

```python
#3d scatter plot

from sklearn.decomposition import PCA

# 3D Scatter Plot of three variables: BMI, Blood Pressure (bp), and Disease Progression (target)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(df['bmi'], df['bp'], df['Disease Progression'], c=df['Disease Progression'], cmap='viridis', s=50)

ax.set_xlabel('BMI')
ax.set_ylabel('Blood Pressure (BP)')
ax.set_zlabel('Disease Progression')
plt.title('3D Scatter Plot of BMI, BP, and Disease Progression')
plt.colorbar(sc, label='Disease Progression')
plt.show()

```
